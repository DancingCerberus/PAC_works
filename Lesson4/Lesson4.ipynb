1. Отсортировать значения массива по частоте встречания.

```python
a = np.random.randint(5, size=10)

unique_elements, frequency = np.unique(a, return_counts=True)
sorted_indexes = np.argsort(frequency)[::-1]
sorted_by_freq = unique_elements[sorted_indexes]

print('Исходный:', a)
print('Уникальные значения:', unique_elements)
print('Частоты:', sorted_indexes)
print('Отсортированный по частоте:', sorted_by_freq)

```
